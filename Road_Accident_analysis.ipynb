```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/RoadAccidents/Road_Accidents_as_per_the_Time_of_Occurrence.csv")

# Filter data for FY 2018 to FY 2022
filtered_data = data[(data['YearCode'] >= 2018) & (data['YearCode'] <= 2022)]

print(filtered_data.head())
```

```python
cleaned_data = data.dropna(subset=['Road accidents occurred in Day or Night'])
accident_by_time = cleaned_data.groupby('Road accidents occurred in Day or Night')['Number of road accidents as per the time of occurrence'].sum()

#plotting
accident_by_time.plot(kind='bar',color=['blue','orange'])
plt.xlabel('Time of occurance')
plt.ylabel('Total number of accidents')
plt.title('Road accidents according Day and night ')

plt.show()
```

```python
unique_class = data['Time of occurrence in Hours'].unique()
print(unique_class)
```

```python
filter_data = data[data['Time of occurrence in Hours'].isin(['06-900hrs','09-1200hrs','12-1500hrs','15-1800hrs','18-2100hrs','21-2400hrs','00-300hrs','03-600hrs','Unknown Time'])]

#group by
accident_by_time = filter_data.groupby('Time of occurrence in Hours')['Number of road accidents as per the time of occurrence'].sum()
print(accident_by_time)

#plot bar graph
accident_by_time.plot(kind = 'bar',color = ['violet','b','g','y','orange','r','purple','black','skyblue'])
plt.xlabel('Time of occurance')
plt.ylabel('Total number of accidents')
plt.title("Number of accidents according to hrs")
plt.show()
```
